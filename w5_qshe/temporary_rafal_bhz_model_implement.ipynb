# Quantum spin Hall effect

```python
import scipy.linalg as la
```

```python
sigma0 = np.array([[1, 0], [0, 1]])
sigmax = np.array([[0, 1], [1, 0]])
sigmay = np.array([[0, -1j], [1j, 0]])
sigmaz = np.array([[1, 0], [0, -1]])
```

We take parameters from C. Liu and S.-C. Zhang, in "Contemporary Concepts of Condensed Matter
Science", Topological Insulators, Vol. 6, edited by Marcel Franz and Laurens
Molenkamp (Elsevier, 2013) pp.59-89:


```python
bhz_parameters = {
    "inverted" : { "A": 3.65, "B": -68.6, "D": -51.1, "M": -0.01 },
    "trivial"  : { "A": 3.87, "B": -48.0, "D": -30.6, "M": +0.009 }
}
```

```python
def get_theoretical_dispersion(kx, ky, par):
    A, B, D, M = par.A, par.B, par.D, par.M
    tmp = np.sqrt(A**2*(kx**2+ky**2) + (M-B*(kx**2+ky**2))**2)
    return -D * (kx**2+ky**2) - tmp, -D * (kx**2+ky**2) + tmp


def onsite(site, par):
    a = 10 * par.a # scaling lattice constant in angstroms (parameters purpose)
    A, B, D, M = par.A/(2*a), par.B/a**2, par.D/a**2, par.M
    return (M - 4 * B) * np.kron(sigma0, sigmaz) - 4 * D * np.kron(sigma0, sigma0)


def hopx(site1, site2, par):
    a = 10 * par.a # scaling lattice constant in angstroms (parameters purpose)
    A, B, D, M = par.A/(2*a), par.B/a**2, par.D/a**2, par.M
    return B * np.kron(sigma0, sigmaz) + D * np.kron(sigma0, sigma0) + 1j * A * np.kron(sigmaz, sigmax)


def hopy(site1, site2, par):
    a = 10 * par.a # scaling lattice constant in angstroms (parameters purpose) 
    A, B, D, M = par.A/(2*a), par.B/a**2, par.D/a**2, par.M
    return B * np.kron(sigma0, sigmaz) + D * np.kron(sigma0, sigma0) - 1j * A * np.kron(sigma0, sigmay)
    
    
def make_bulk(a=1):
    a = 10 * a    # scaling lattice constant in angstroms (parameters purpose)

    def func(site, par):
        kx,ky = par.kx, par.ky
        H0, H1, H2 = onsite(site, par), hopx(site, site, par), hopy(site, site, par)
        
        return H0 \
            + H1*np.exp(-1j*kx*a) + H1.transpose().conjugate()*np.exp(+1j*kx*a) \
            + H2*np.exp(-1j*ky*a) + H2.transpose().conjugate()*np.exp(+1j*ky*a)
    
    lat = kwant.lattice.square()
    sys = kwant.Builder()
    sys[lat(0, 0)] = func

    return sys.finalized()


def make_ribbon(w=100, a=1):
    def ribbon_shape(pos):
        (x, y) = pos
        return (0 <= y < w )
    
    lat = kwant.lattice.square(a=a)
    sym = kwant.TranslationalSymmetry((a, 0))
    sys = kwant.Builder(sym)
    
    sys[lat.shape(ribbon_shape, (0, 0))] = onsite
    sys[kwant.HoppingKind((1,0), lat)] = hopx
    sys[kwant.HoppingKind((0,1), lat)] = hopy
    
    return sys.finalized()
```

## Testing bulk

```python
a = 1
sys = make_bulk(a=a)
momenta = np.linspace(-.01, .01, 21)

par = SimpleNamespace(kx=0, ky=0, a=a, **bhz_parameters['inverted'])

E = []
theory = []
for par.kx in momenta:
    matrix = sys.hamiltonian_submatrix(args=[par])
    ev = la.eigh(matrix)[0]
    E.append(ev)
    theory.append(get_theoretical_dispersion(par.kx,0, par))
    
plt.plot(momenta, E, 'k-');
plt.plot(momenta, theory,'ro')

plt.xlim(momenta.min(), momenta.max())
```

## Testing Edge States

```python
a = 2
par = SimpleNamespace(a=a, **bhz_parameters['inverted'])
sys = make_ribbon(w=200, a=a)
```

```python
kwant.plotter.bands(sys, args=[par], momenta=np.linspace(-0.3, 0.3, 100, endpoint=False), show=False);
plt.ylim(-.05,.05)
```
