# Quantum spin Hall effect

```python
import kwant
```

```python
import numpy as np
import scipy.linalg as la
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
from IPython.html.widgets import interact
```

```python
sigma0 = np.array([[1, 0], [0, 1]])
sigmax = np.array([[0, 1], [1, 0]])
sigmay = np.array([[0, -1j], [1j, 0]])
sigmaz = np.array([[1, 0], [0, -1]])
```

We take parameters from C. Liu and S.-C. Zhang, in "Contemporary Concepts of Condensed Matter
Science", Topological Insulators, Vol. 6, edited by Marcel Franz and Laurens
Molenkamp (Elsevier, 2013) pp.59-89:


```python
params = {
    "inverted" : { "A": 3.65, "B": -68.6, "D": -51.1, "M": -0.01 },
    "trivial"  : { "A": 3.87, "B": -48.0, "D": -30.6, "M": +0.009 }
}
```

```python
def get_theoretical_dispersion(kx, ky, material='inverted'):
    mat = params[material]
    A, B, D, M = mat["A"], mat["B"], mat["D"], mat["M"]
    
    tmp = np.sqrt(A**2*(kx**2+ky**2) + (M-B*(kx**2+ky**2))**2)
    return -D * (kx**2+ky**2) - tmp, -D * (kx**2+ky**2) + tmp


def make_bulk(material="inverted", a=10):
    lat = kwant.lattice.square(a)

    mat = params[material]
    A, B, D, M = mat["A"], mat["B"], mat["D"], mat["M"]
    
    a_in_A = 10 * a    # lattice constant in ansgroms
    
    A = A/(2*a_in_A)
    B = B/a_in_A**2
    D = D/a_in_A**2
    
    onsite = (M - 4 * B) * np.kron(sigma0, sigmaz) - 4 * D * np.kron(sigma0, sigma0)
    hopx = B * np.kron(sigma0, sigmaz) + D * np.kron(sigma0, sigma0) + 1j * A * np.kron(sigmaz, sigmax)
    hopy = B * np.kron(sigma0, sigmaz) + D * np.kron(sigma0, sigma0) - 1j * A * np.kron(sigma0, sigmay)
    
    
    def func(site, kx, ky):
        return onsite \
            + hopx*np.exp(-1j*kx*a_in_A) + hopx.transpose().conjugate()*np.exp(+1j*kx*a_in_A) \
            + hopy*np.exp(-1j*ky*a_in_A) + hopy.transpose().conjugate()*np.exp(+1j*ky*a_in_A)
    
    sys = kwant.Builder()
    
    sys[lat(0, 0)] = func

    return sys.finalized()


def make_ribbon(w=100, material="inverted", a=10):
    mat = params[material]
    A, B, D, M = mat["A"], mat["B"], mat["D"], mat["M"]
    a_in_A = 10 * a # lattice constant in ansgroms
    
    A = A/(2*a_in_A)
    B = B/a_in_A**2
    D = D/a_in_A**2
    
    onsite = (M - 4 * B) * np.kron(sigma0, sigmaz) - 4 * D * np.kron(sigma0, sigma0)
    hopx = B * np.kron(sigma0, sigmaz) + D * np.kron(sigma0, sigma0) + 1j * A * np.kron(sigmaz, sigmax)
    hopy = B * np.kron(sigma0, sigmaz) + D * np.kron(sigma0, sigma0) - 1j * A * np.kron(sigma0, sigmay)
    
    def ribbon_shape(pos):
        (x, y) = pos
        return (0 <= y < w )
    
    lat = kwant.lattice.square(a=a)
    sym = kwant.TranslationalSymmetry((a, 0))
    sys = kwant.Builder(sym)
    
    sys[lat.shape(ribbon_shape, (0, 0))] = onsite
    sys[kwant.HoppingKind((1,0), lat)] = hopx
    sys[kwant.HoppingKind((0,1), lat)] = hopy
    
    return sys.finalized()
```

## Testing bulk

```python
a = 1
sys = make_bulk(a=a)
momenta = np.linspace(-.01, .01, 21)

E = []
theory = []
for k in momenta:
    matrix = sys.hamiltonian_submatrix(args=[k,0])
    ev = la.eigh(matrix)[0]
    E.append(ev)
    theory.append(get_theoretical_dispersion(k,0))
    
plt.plot(momenta, E, 'k-');
plt.plot(momenta, theory,'ro')

plt.xlim(momenta.min(), momenta.max())
```

## Testing Edge States

```python
sys = make_ribbon(w=200, a=2)
```

```python
kwant.plotter.bands(sys, momenta=np.linspace(-0.3, 0.3, 100, endpoint=False), show=False);
plt.ylim(-.05,.05)
```

## Testing trivial edge states

```python
sys = make_ribbon(w=200, a=2, material='trivial')
```

```python
kwant.plotter.bands(sys, momenta=np.linspace(-0.3, 0.3, 100, endpoint=False), show=False);
plt.ylim(-.05,.05)
```
