```python
%run ../code/init_mooc_nb.ipy

import scipy.linalg as sla
from matplotlib import cm

sigma0 = np.array([[1, 0], [0, 1]])
sigmax = np.array([[0, 1], [1, 0]])
sigmay = np.array([[0, -1j], [1j, 0]])
sigmaz = np.array([[1, 0], [0, -1]])
```

```python
def warping_energies(par):
    def func(kx, ky):
        k = np.sqrt(kx**2+ky**2)
        matrix = par.t * k**2 * sigma0  \
                + par.v * (1+par.alpha * k**2) * (kx * sigmay - ky * sigmax) \
                + 0.5 * par.lamb * ( (kx+1j*ky)**3 + (kx-1j*ky)**3 ) * sigmaz
        ev = sla.eigh(matrix)[0]
        return ev
    return func


```

```python
def evaluate_on_grid(X, Y, func):
    """ X, Y should be in np.meshgrid form. It's enough for func to work on floats. """
    data = []
    for xx, yy in zip(X, Y):
        row = []
        for i,j in zip(xx, yy):
            row.append(func(i,j))
        data.append(row)
    data = np.array(data)
    data = [np.array(data[:,:,i]) for i in range(np.shape(data)[2])]
    return data


def func(kx, ky):
    matrix = (kx * sigmay - ky * sigmax) + 0.5 * ( (kx+1j*ky)**3 + (kx-1j*ky)**3 ) * sigmaz
    return sla.eigh(matrix)[0]


def plot_warping(Kmax=0.2):
#     K = np.linspace(-Kmax, Kmax, 101)
    
    N = 100
    r = np.linspace(0,Kmax,N)
    p = np.linspace(0,2*np.pi,N)
    
    R,P = np.meshgrid(r,p)
    # transform them to cartesian system
    X,Y = R*np.cos(P),R*np.sin(P)

    energies = evaluate_on_grid(X, Y, func=func)

    fig = plt.figure(figsize=(12, 8))
    ax = fig.add_subplot(111, projection='3d')
    
    Z = energies
    vmin = np.array(Z).min()
    vmax = np.array(Z).max()
    
    for z in energies:
        ax.plot_surface(X, Y, z, rstride=1, cstride=1, cmap=cm.RdBu_r, linewidth=0.05, vmin=vmin, vmax=vmax)
    
    ax.set_xlabel('$k_x$')# [\sqrt{v/\lambda}]$')
    ax.set_ylabel('$k_y$')# [\sqrt{v/\lambda}]$')
    ax.set_zlabel('$E$')# [E^{*}]$')
    
    ax.set_xlim3d(-1.4*Kmax, 1.4*Kmax)
    ax.set_ylim3d(-1.4*Kmax, 1.4*Kmax)
    
    vals = [-1.5, 0.0, 1.5]
    ax.set_xticks(vals)
    ax.set_xticklabels(["${0}$".format(i) for i in vals])

    ax.set_yticks(vals)
    ax.set_yticklabels(["${0}$".format(i) for i in vals])

    ax.set_zticks(vals)
    ax.set_zticklabels(["${0}$".format(i) for i in vals])
    
#     ax.view_init(8,20)
    ax.set_zlim3d(-1.5, 1.5)

```
