```python
def nanowire_chains(length=20, n=2):
    
    def onsite(site, par):
        (x, y) = site.pos
        return ((2*par.t + 2*par.ty - par.mu) * np.kron(sigmaz, sigma0) + par.delta * np.kron(sigmax, sigma0)
                + (x%2==0) * par.B * np.kron(sigma0, sigmaz) + (x%2==1) * par.B * np.kron(sigma0, sigmay))

    def hopping(site1, site2, par):
        return -par.t * np.kron(sigmaz, sigma0) + 0.5 * 1j * par.alpha * np.kron(sigmaz, sigmax)
    
    def shape(pos):
        (x, y) = pos
        return (0 <= x < length) * (0 <= y < n)
        
    lat = kwant.lattice.square()
    sym = kwant.TranslationalSymmetry((0, n))
    sys = kwant.Builder(sym)    
    
    sys[lat.shape(shape, (0, 0))] = onsite
    sys[kwant.HoppingKind((1, 0), lat)] = hopping
    sys[kwant.HoppingKind((0, 1), lat)] = -par.ty * np.kron(sigmaz, sigma0)
    
    return sys.finalized()
```

```python
def qshes(r=5, n=2):
    def shape(pos):
        (x, y, z) = pos
        return ((x**2 + y**2) < r**2) 

    def onsite(site, par):
        A, B, D, M = par.A/2, par.B, par.D, par.M
        return (M - 4 * B) * np.kron(sigma0, sigmaz) - 4 * D * np.kron(sigma0, sigma0) \
            + par.del_z * np.kron(sigmay, sigmay)

    def hopx(site1, site2, par):
        A, B, D, M = par.A/2, par.B, par.D, par.M
        return B * np.kron(sigma0, sigmaz) + D * np.kron(sigma0, sigma0) + 1j * A * np.kron(sigmaz, sigmax)

    def hopy(site1, site2, par):
        A, B, D, M = par.A/2, par.B, par.D, par.M
        return B * np.kron(sigma0, sigmaz) + D * np.kron(sigma0, sigma0) - 1j * A * np.kron(sigma0, sigmay)
    
    lat = kwant.lattice.general(np.identity(3))
    sym = kwant.TranslationalSymmetry((0, 0, n))
    sys = kwant.Builder(sym)

    sys[lat.shape(shape, (0, 0, 0))] = onsite
    sys[kwant.HoppingKind((1, 0, 0), lat)] = hopx
    sys[kwant.HoppingKind((0, 1, 0), lat)] = hopy
    sys[kwant.HoppingKind((0, 0, 1), lat)] = par.tz * np.kron(sigma0, sigmaz)
    return sys.finalized()
```

```python
par = SimpleNamespace(A=1.0, B=1.0, D=0.0, M=0.2, del_z=0.0, tz=0.0)
sys = qshes()

kwant.plotter.bands(sys, args=[par,], show=False)
plt.ylim(-0.4, 0.4)
```
