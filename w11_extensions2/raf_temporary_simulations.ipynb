$$
H = (\epsilon_p - \mu) \tau_z \otimes \sigma_0 + \alpha p \, \tau_z \otimes \sigma_x + B \, \tau_0 \otimes \sigma_x + \Delta \, \tau_x \otimes \sigma_0
$$

```python
sigma0 = np.array([[1., 0.], [0., 1.]])
sigmax = np.array([[0., 1.], [1., 0.]])
sigmay = np.array([[0., -1j], [1j, 0.]])
sigmaz = np.array([[1., 0.], [0., -1.]])

def onsite(site, par):
    return ((2*par.t - par.mu) * np.kron(sigmaz, sigma0) + par.B * np.kron(sigma0, sigmaz) 
            + par.delta * np.kron(sigmax, sigma0))

def hopping(site1, site2, par):
    return -par.t * np.kron(sigmaz, sigma0) + 0.5 * 1j * par.alpha * np.kron(sigmaz, sigmax)
```

```python
def nanowire_chain(L=10):
    lat = kwant.lattice.chain()
    sys = kwant.Builder()    
    
    sys[(lat(x) for x in xrange(L))] = onsite
    sys[kwant.HoppingKind((1,), lat)] = hopping
    
    lead = kwant.Builder(kwant.TranslationalSymmetry((-1,)))
    lead[lat(0)] = onsite
    lead[kwant.HoppingKind((1,), lat)] = hopping
    
    sys.attach_lead(lead)
    sys.attach_lead(lead.reversed())
    
    return sys.finalized()
```

```python
sys = nanowire_chain(10)
```

```python
J = 2.

par1 = SimpleNamespace(t=J/2, mu=-1*J, B=J, delta=2*J, alpha=J)
par2 = SimpleNamespace(t=J/2, mu=-4*J, B=J, delta=2*J, alpha=J)


def get_evolution_operator_finite(sys, J, T):
    H1 = sys.hamiltonian_submatrix(args=[par1])
    H2 = sys.hamiltonian_submatrix(args=[par2])
    
    exp1 = la.expm(-1j*H1 * 0.5 * T)
    exp2 = la.expm(-1j*H2 * 0.5 * T)
    
    return np.dot(exp1,exp2)
```

```python
sys = nanowire_chain(25)

periods = np.linspace(0.3/J, 1.6/J, 101)

energies = []
for T in periods:
    U = get_evolution_operator_finite(sys, J, T)
    phases = ((1j) * np.log( la.eig(U)[0] )).real
    
    phases = np.sort(np.abs(phases))
    phases = np.sort([(-1)**(n) * val for n, val in enumerate(phases)])
    
    energies.append(phases)
    
energies = np.array(energies)
```

```python
plt.figure(figsize=(8,6))
plt.plot(periods*J, energies.real, 'k-');
```
