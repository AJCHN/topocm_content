```python
t = 1.0
tprime = 1/np.sqrt(2)
fi = np.pi/4
EpsA = 0.0
EpsB = 0.0
```

```python
onsite = np.array([[EpsA, t*np.exp(-1j*fi)], [t*np.exp(+1*fi), EpsB]], dtype=complex)

hopping_x = np.array([[-tprime, t*np.exp(+1j*fi)], [0.0, +tprime]], dtype=complex)
hopping_y = np.array([[+tprime, t*np.exp(+1j*fi)], [0.0, -tprime]], dtype=complex)

hopping_xy = np.array([[0, t*np.exp(-1j*fi)],[0.0, 0.0]], dtype=complex)
```

```python
hopping_x.transpose().conj()
```

```python
def dag(x):
    return x.transpose().conj()
```

```python
def get_hamiltonian(kx, ky):
    return onsite + (hopping_x*np.exp(-1j*kx) + dag(hopping_x)*np.exp(+1j*kx)) + \
            (hopping_y*np.exp(-1j*ky) + dag(hopping_y)*np.exp(+1j*ky)) + \
            (hopping_xy*np.exp(-1j*(kx+ky)) + dag(hopping_xy)*np.exp(+1j*(kx+ky)))
```

```python
def Eps(kx, ky):
    mat = np.zeros((2,2), dtype=complex)
    
    mat[0,0] = EpsA + 2*tprime*(np.cos(ky)-np.cos(kx))
    mat[1,1] = EpsB + 2*tprime*(np.cos(kx)-np.cos(ky))
#     mat[0,1] = np.exp(-1j*0.5*(kx+ky)) * (2*t*np.exp(-1j*fi) * np.cos(0.5*(kx+ky)) + 2*t*np.exp(+1j*fi) * np.cos(0.5*(kx-ky)))
    mat[0,1] = t * np.exp(-1j*fi) + t*np.exp(1j*fi - 1j*ky) + t*np.exp(1j*fi - 1j*kx) + t*np.exp(-1j*fi - 1j*kx - 1j*ky) 
    mat[1,0] = mat[0,1].conjugate()
    
    e = np.linalg.eigh(mat)[0]
    e = sorted(e)
    return e                                         
```

```python
def get_energy(kx, ky):
    mat = get_hamiltonian(kx,ky)
    e = np.linalg.eigh(mat)[0]
    e = sorted(e)
    return e                                         
```

```python
def interpolate_bz(KX, KY, energies):
    X, Y, Z1, Z2 = [], [], [], []
    
    for kx, ky, e in zip(KX, KY, energies):
        X.append(kx)
        Y.append(ky)
        Z1.append(e[0])
        Z2.append(e[1])

    X = np.array(X)
    Y = np.array(Y)
    Z1 = np.array(Z1)
    Z2 = np.array(Z2)
    
    rbf1 = scipy.interpolate.Rbf(X,Y,Z1, function='linear')
    rbf2 = scipy.interpolate.Rbf(X,Y,Z2, function='linear')
    
    xi, yi = np.linspace(X.min(), X.max(), 50), np.linspace(Y.min(), Y.max(), 50)
    xi, yi = np.meshgrid(xi, yi)

    zi1 = rbf1(xi,yi)
    zi2 = rbf2(xi,yi)
    return xi, yi, [zi1, zi2], rbf1, rbf2
```

```python
def plot_two(X,Y,Z, angle):
    fig = plt.figure(figsize=(14,10))
    ax = fig.add_subplot(111, projection='3d')

    ax.plot_surface(X, Y, Z[0], rstride=1, cstride=1, cmap=cm.YlGnBu_r)
    ax.plot_wireframe(X, Y, Z[1], rstride=1, cstride=1, cmap=cm.YlGnBu_r)

    ax.set_xlabel('kx')
    ax.set_ylabel('ky')

#     ax.set_xlim3d(-.4,.4)
#     ax.set_ylim3d(-.4,.4)
#     ax.set_zlim3d(60, 110)   
    
    ax.view_init(*angle)   
```

```python
def nx(kx, ky):
    return -4 * t * np.cos(fi) * np.cos(0.5*kx) * np.cos(0.5*ky)


def ny(kx, ky):
    return +4 * t * np.sin(fi) * np.sin(0.5*kx) * np.sin(0.5*ky)


def nz(kx, ky):
    return 2 * tprime * (np.cos(ky) - np.cos(kx))

def theory(kx,ky):
    e = nx(kx, ky)**2 + ny(kx, ky)**2 + nz(kx, ky)**2
    e = np.sqrt(e)
    return [-e, e]
```

```python
def theory(kx,ky):
    e = nx(kx, ky)**2 + ny(kx, ky)**2 + nz(kx, ky)**2
    e = np.sqrt(e)
    return [-e, e]
```
