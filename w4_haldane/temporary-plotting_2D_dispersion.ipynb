## Function for dispersion plotting

```python
def evaluate_on_grid(X, Y, func):
    """ X, Y should be in np.meshgrid form. It's enough for func to work on floats. """
    data = []
    for xx, yy in zip(KX, KY):
        row = []
        for i,j in zip(xx, yy):
            row.append(func(i,j))
        data.append(row)
    data = np.array(data)
    data = [np.array(data[:,:,i]) for i in range(np.shape(data)[2])]
    return data
```

```python
def get_energies(sys, par):
    mat = sys.hamiltonian_submatrix(args=[par])
    e = np.linalg.eigh(mat)[0]
    return sorted(e)
```

```python
def dispersion_func(sys, par):
    def func(kx, ky):
        par.kx = kx
        par.ky = ky
        return get_energies(sys, par)
    return func       
```

```python
def plot_two(X,Y,Z, angle):
    fig = plt.figure(figsize=(14,10))
    ax = fig.add_subplot(111, projection='3d')

    ax.plot_surface(X, Y, Z[0], rstride=1, cstride=1, cmap=cm.YlGnBu_r)
    ax.plot_wireframe(X, Y, Z[1], rstride=1, cstride=1, cmap=cm.YlGnBu_r)

    ax.set_xlabel('kx')
    ax.set_ylabel('ky')

#     ax.set_xlim3d(-.4,.4)
#     ax.set_ylim3d(-.4,.4)
#     ax.set_zlim3d(60, 110)   
    
    ax.view_init(*angle)   
```
