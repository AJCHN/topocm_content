```python
%run ../code/init_mooc_nb.ipy

class SimpleNamespace(object):
    def __init__(self, **kwargs):
        self.__dict__.update(kwargs)
        
        
def plot_1D_wfsqr(sys, evec, n):
    x = [sys.site(i).pos[0] for i in range(sys.graph.num_nodes)]
    x = np.array(x)
    y = np.abs(evec[:,n])**2
    indx = np.argsort(x)
   
    plt.plot(x[indx],y[indx],'-')
    return x[indx]


def modulated_wire(L=50):
    """Create a modulated wire. 
    
    Chain lattice, one orbital per site.
    Returns finalized kwant system.
    
    Arguments required in onsite/hoppings: 
        t, mu, mu_lead, A, omega, phase
    """
    def onsite(site, par):
        x = site.pos[0]
        
        return 2*par.t - par.mu + par.A * np.cos(par.omega * x + par.phase) 
    
    def hopping(target, source, par):
        return -par.t
    
    lat = kwant.lattice.chain()
    sys = kwant.Builder()
    
    sys[(lat(x) for x in range(L))] = onsite
    sys[lat.neighbors()] = hopping
    
    sym_lead = kwant.TranslationalSymmetry([-1])
    lead = kwant.Builder(sym_lead)
    
    lead[lat(0)] = lambda site, par: 2*par.t - par.mu_lead
    lead[lat.neighbors()] = lambda target, source, par: -par.t
    
    sys.attach_lead(lead)
    sys.attach_lead(lead.reversed())
    
    return sys.finalized()


def qhe_corbino(r_out=20, r_in=10, w_lead=10):
    """Create corbino disk. 
    
    Square lattice, one orbital per site.
    Returns finalized kwant system.
    
    Arguments required in onsite/hoppings: 
        t, mu, mu_lead, B, phi
    """
    # ring shape
    def ring(pos):
        (x, y) = pos
        rsq = x ** 2 + y ** 2
        return (r_in ** 2 < rsq < r_out ** 2)

    # Onsite and hoppings
    def onsite(site, par):
        (x,y) = site.pos

        return 4*par.t - par.mu   
    
    def crosses_branchcut(hop):
        xt, yt = hop[0].pos
        xs, ys = hop[1].pos
        return yt < 0 and xt > 0.5 and xs < 0.5 

    def hopping(target, source, par):
        xt, yt = target.pos
        xs, ys = source.pos

        return -par.t * np.exp(-0.5j * par.B * (xt - xs) * (yt + ys))
    
    def branchcut_hopping(target, source, par):
        return hopping(target, source, par) * np.exp(1j * par.phi)
            
    # Building system
    lat = kwant.lattice.square()
    sys = kwant.Builder()
    
    sys[lat.shape(ring, (0, r_in + 1))] = onsite
    sys[lat.neighbors()] = hopping 
    
    # adding special hoppings
    def hops_across_cut(sys):
        for hop in kwant.builder.HoppingKind((1, 0), lat, lat)(sys):
            if crosses_branchcut(hop):
                yield hop
    sys[hops_across_cut] = branchcut_hopping
    
    # Attaching leads
    sym_lead = kwant.TranslationalSymmetry((-1, 0))
    lead = kwant.Builder(sym_lead)

    def lead_shape(pos):
        (x, y) = pos
        return (-w_lead / 2 < y < w_lead / 2)

    lead[lat.shape(lead_shape, (0, 0))] = lambda site, par: 4 * par.t - par.mu_lead
    lead[lat.neighbors()] = lambda target, source, par: -par.t

    #### Attach the leads and return the system. ####
    sys.attach_lead(lead)
    sys.attach_lead(lead, origin=lat(0,0))

    return sys.finalized()

def qhe_hall_bar(L=50, W=10, w_lead=10):
    """Create a hall bar system. 
    
    Square lattice, one orbital per site.
    Returns finalized kwant system.
    
    Arguments required in onsite/hoppings: 
        t, mu, mu_lead
    """
    # bar shape
    def bar(pos):
        (x, y) = pos
        return (x > -L/2 and x < L/2) and (y > -W/2 and y < W/2)

    # Onsite and hoppings
    def onsite(site, par):
        (x,y) = site.pos

        return 4*par.t - par.mu   
    
    def hopping(target, source, par):
        xt, yt = target.pos
        xs, ys = source.pos

        return -par.t * np.exp(-0.5j * par.B * (xt - xs) * (yt + ys))
            
    # Building system
    lat = kwant.lattice.square()
    sys = kwant.Builder()
    
    sys[lat.shape(bar, (0, 0))] = onsite
    sys[lat.neighbors()] = hopping 
    
    # Attaching leads
    sym_lead = kwant.TranslationalSymmetry((-1, 0))
    lead = kwant.Builder(sym_lead)

    def lead_shape(pos):
        (x, y) = pos
        return (-w_lead / 2 < y < w_lead / 2)

    lead[lat.shape(lead_shape, (-1, 0))] = lambda site, par: 4 * par.t - par.mu_lead
    lead[lat.neighbors()] = lambda target, source, par: -par.t

    #### Attach the leads and return the system. ####
    sys.attach_lead(lead)
    sys.attach_lead(lead.reversed())

    return sys.finalized()
```

```python
def plot_conductance(sys, par, energy, fluxes):
    """ Compute conductance. """
    normalized_fluxes = [flux / (2 * np.pi) for flux in fluxes]
    data = []
    for flux in fluxes:
        par.phi=flux
        smatrix = kwant.smatrix(sys, energy, args=[par])
        data.append(smatrix.transmission(1, 0))
 
    plt.figure()
    plt.plot(normalized_fluxes, data)
    plt.xlabel("flux [flux quantum]")
    plt.ylabel("conductance [e^2/h]")
    plt.show()
plot_conductance(sys, par, energy=0.15, fluxes=fluxes)
```
