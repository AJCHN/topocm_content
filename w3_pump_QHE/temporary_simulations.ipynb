```python
%run ../code/init_mooc_nb.ipy

class SimpleNamespace(object):
    def __init__(self, **kwargs):
        self.__dict__.update(kwargs)
        
        
def plot_1D_wfsqr(sys, evec, n):
    x = [sys.site(i).pos[0] for i in range(sys.graph.num_nodes)]
    x = np.array(x)
    y = np.abs(evec[:,n])**2
    indx = np.argsort(x)
   
    plt.plot(x[indx],y[indx],'-')
    return x[indx]


def modulated_wire(L=50):
    """Create a modulated wire. 
    
    Chain lattice, one orbital per site.
    Returns finalized kwant system.
    
    Arguments required in onsite/hoppings: 
        t, mu, mu_lead, A, omega, phase
    """
    def onsite(site, par):
        x = site.pos[0]
        
        return 2*par.t - par.mu + par.A * np.cos(par.omega * x + par.phase) 
    
    def hopping(target, source, par):
        return -par.t
    
    lat = kwant.lattice.chain()
    sys = kwant.Builder()
    
    sys[(lat(x) for x in range(L))] = onsite
    sys[lat.neighbors()] = hopping
    
    sym_lead = kwant.TranslationalSymmetry([-1])
    lead = kwant.Builder(sym_lead)
    
    lead[lat(0)] = lambda site, par: 2*par.t - par.mu_lead
    lead[lat.neighbors()] = lambda target, source, par: -par.t
    
    sys.attach_lead(lead)
    sys.attach_lead(lead.reversed())
    
    return sys.finalized()


def qhe_corbino(r_out=100, r_in=65, w_lead=10):
    """Create corbino disk. 
    
    Square lattice, one orbital per site.
    Returns finalized kwant system.
    
    Arguments required in onsite/hoppings: 
        t, mu, mu_lead, B, phi
    """
    # ring shape
    def ring(pos):
        (x, y) = pos
        rsq = x ** 2 + y ** 2
        return (r_in ** 2 < rsq < r_out ** 2)

    # Onsite and hoppings
    def onsite(site, par):
        (x,y) = site.pos
        return 4*par.t - par.mu   
    
    def crosses_branchcut(hop):
        xt, yt = hop[0].pos
        xs, ys = hop[1].pos
        return yt < 0 and xt > 0.5 and xs < 0.5 

    def hopping(target, source, par):
        xt, yt = target.pos
        xs, ys = source.pos
        # Check for correctness!
        return -par.t * np.exp(-0.5j * par.B * (xt - xs) * (yt + ys))
    
    def branchcut_hopping(target, source, par):
        return hopping(target, source, par) * np.exp(1j * par.phi)
            
    # Building system
    lat = kwant.lattice.square()
    sys = kwant.Builder()
    
    sys[lat.shape(ring, (0, r_in + 1))] = onsite
    sys[lat.neighbors()] = hopping 
    
    # adding special hoppings
    def hops_across_cut(sys):
        for hop in kwant.builder.HoppingKind((1, 0), lat, lat)(sys):
            if crosses_branchcut(hop):
                yield hop
    sys[hops_across_cut] = branchcut_hopping
    
    # Attaching leads
    sym_lead = kwant.TranslationalSymmetry((-1, 0))
    lead = kwant.Builder(sym_lead)

    def lead_shape(pos):
        (x, y) = pos
        return (-w_lead / 2 < y < w_lead / 2)

    lead[lat.shape(lead_shape, (0, 0))] = lambda site, par: 4 * par.t - par.mu_lead
    lead[lat.neighbors()] = lambda target, source, par: -par.t

    #### Attach the leads and return the system. ####
    sys.attach_lead(lead)
    sys.attach_lead(lead, origin=lat(0,0))

    return sys.finalized()

def total_charge(value_array):
    determinants = [np.linalg.det(s) for s in value_array]
    charge = np.cumsum(np.angle(np.roll(determinants, -1) / determinants))
    return charge / (2*np.pi)


def qhe_hall_bar(L=50, W=10, w_lead=10):
    """Create a hall bar system. 
    
    Square lattice, one orbital per site.
    Returns finalized kwant system.
    
    Arguments required in onsite/hoppings: 
        t, mu, mu_lead
    """
    # bar shape
    def bar(pos):
        (x, y) = pos
        return (x > -L/2 and x < L/2) and (y > -W/2 and y < W/2)

    # Onsite and hoppings
    def onsite(site, par):
        (x,y) = site.pos
        return 4*par.t - par.mu   
    
    def hopping(target, source, par):
        xt, yt = target.pos
        xs, ys = source.pos
        return -par.t * np.exp(-0.5j * par.B * (xt - xs) * (yt + ys))
            
    # Building system
    lat = kwant.lattice.square()
    sys = kwant.Builder()
    
    sys[lat.shape(bar, (0, 0))] = onsite
    sys[lat.neighbors()] = hopping 
    
    # Attaching leads
    sym_lead = kwant.TranslationalSymmetry((-1, 0))
    lead = kwant.Builder(sym_lead)

    def lead_shape(pos):
        (x, y) = pos
        return (-w_lead / 2 < y < w_lead / 2)

    lead[lat.shape(lead_shape, (-1, 0))] = lambda site, par: 4 * par.t - par.mu_lead
    lead[lat.neighbors()] = lambda target, source, par: -par.t

    #### Attach the leads and return the system. ####
    sys.attach_lead(lead)
    sys.attach_lead(lead.reversed())

    sym_lead_vertical = kwant.TranslationalSymmetry((0,1))
    lead_vertical1 = kwant.Builder(sym_lead_vertical)
    lead_vertical2 = kwant.Builder(sym_lead_vertical)
    
    def lead_shape_vertical1(pos):
        (x, y) = pos
        return (-w_lead - 10 < x < - 10)
    
    def lead_shape_vertical2(pos):
        (x, y) = pos
        return (10 < x < w_lead + 10)
    
    lead_vertical1[lat.shape(lead_shape_vertical1, (-11, 0))] = lambda site, par: 4 * par.t - par.mu_lead
    lead_vertical1[lat.neighbors()] = lambda target, source, par: -par.t
    lead_vertical2[lat.shape(lead_shape_vertical2, (11, 0))] = lambda site, par: 4 * par.t - par.mu_lead
    lead_vertical2[lat.neighbors()] = lambda target, source, par: -par.t
     
    sys.attach_lead(lead_vertical1)
    sys.attach_lead(lead_vertical1.reversed())
    sys.attach_lead(lead_vertical2)
    sys.attach_lead(lead_vertical2.reversed())
    
    return sys.finalized()

def winding_number(value_array):
    determinants = [ np.linalg.det(s) for s in value_array]
    winding = np.sum(np.angle(np.roll(determinants, -1) / determinants))
    return np.round(winding/(2*np.pi))

def qhe_ribbon(W, periodic=False):
    """ Creates blabla bla system
    
    If we have periodic boundary conditions, the flux through a single 
    unit cell is quantized.
    """
    W = 2*(W//2)
    
    def ribbon_shape(pos):
        (x, y) = pos
        return (-W / 2 <= y <= W / 2)
    
    def onsite(site, par):
        (x,y) = site.pos
        return 4*par.t - par.mu
    
    def hopping(target, source, par):
        xt, yt = target.pos
        xs, ys = source.pos
        return -par.t * np.exp(-0.5j * par.B * (xt - xs) * (yt + ys))
    
    def hopping_periodic(target, source, par):
        xt, yt = target.pos
        xs, ys = source.pos
        return -par.t * np.exp(-1j * np.pi / (W + 1) * np.round((W + 1) * par.B /(2 * np.pi)) * (xt - xs) * (yt + ys))        
    
    lat = kwant.lattice.square()
    sym_sys = kwant.TranslationalSymmetry((-1, 0))
    sys = kwant.Builder(sym_sys)
    
    sys[lat.shape(ribbon_shape, (0, 0))] = onsite
    
    if periodic:
        sys[lat.neighbors()] = hopping_periodic
        sys[lat(0, - W / 2 ), lat(0, + W / 2)] = hopping_periodic
    else:
        sys[lat.neighbors()] = hopping
    
    return sys.finalized()

def sigma(s):
    # reduce by one dimension G -> G[temp, temp]
    temp = [0, 1, 3, 4, 5]
    s = s[temp, :]
    s = s[:, temp]
    # invert R = G^-1
    # find out whether it is a numpy object
    r = np.linalg.inv(s)
    # Voltages follow: V = R I[temp]
    V = r.dot(np.array([0, 0, 0, 1, -1]))
    # Completely solved the six terminal system. 
    # Consider the 2x2 conductance now: Use I = sigma U
    E_x = V[0] - V[1] - V[3]
    E_y = V[0] + V[1] - V[3]
    #formula above
    sigma_xx = E_x / (E_x**2 + E_y**2)
    sigma_xy = E_y / (E_x**2 + E_y**2)
    return sigma_xx, sigma_xy    
```

##modulated nanowire

$$H = -t \sum_i c_{i+1}^\dagger c_i - \sum_i(2t-\mu) c_i^\dagger c_i + \sum_i U(x) c_i^\dagger c_i$$

$$ U(x) = A \cos(\omega x + \phi) $$



## corbino disk

$$ H = -t \sum_{<i,j>} e^{i \phi_{ij}} c_i^\dagger c_j - (4t-\mu) c_i^\dagger c_i $$

$$ \phi_{ij} = -\frac{B}{2}  (x_i - x_j)(y_i+y_j) $$

```python
def plot_conductance(sys, par, energy, fluxes):
    """ Compute conductance. """
    normalized_fluxes = [flux / (2 * np.pi) for flux in fluxes]
    data = []
    for flux in fluxes:
        par.phi=flux
        smatrix = kwant.smatrix(sys, energy, args=[par])
        data.append(smatrix.transmission(1, 0))
 
    plt.figure()
    plt.plot(normalized_fluxes, data)
    plt.xlabel("flux [flux quantum]")
    plt.ylabel("conductance [e^2/h]")
    plt.show()
plot_conductance(sys, par, energy=0.15, fluxes=fluxes)
```
