```python
import phlead
sigma0 = np.array([[1., 0.], [0., 1.]])
sigmax = np.array([[0., 1.], [1., 0.]])
sigmay = np.array([[0., -1j], [1j, 0.]])
sigmaz = np.array([[1., 0.], [0., -1.]])



### Tunel spectroscopy
def make_system_spectroscopy(t=1):
    def onsite_sc(site, mu_l, mu_sc, Ez, alpha, delta, Vbarrier):
        return (2*t - mu_sc) * np.kron(sigma0, sigmaz) + Ez * np.kron(sigmax, sigma0) + delta * np.kron(sigma0, sigmax)

    def onsite_normal(site, mu_l, mu_sc, Ez, alpha, delta, Vbarrier):
        return (2*t - mu_l) * np.kron(sigma0, sigmaz) + Ez * np.kron(sigmax, sigma0)
    
    def onsite_barrier(site, mu_l, mu_sc, Ez, alpha, delta, Vbarrier):
        return (2*t - mu_l + Vbarrier) * np.kron(sigma0, sigmaz) + Ez * np.kron(sigmax, sigma0) 
    
    def hop(site1, site2, mu_l, mu_sc, Ez, alpha, delta, Vbarrier):
        return -t * np.kron(sigma0, sigmaz) + 1j * alpha * np.kron(sigmay, sigmaz)
    
    lat = kwant.lattice.chain()
    sys = kwant.Builder()
    
    # The system just consists of the tunnel barrier
    sys[lat(0)] = onsite_barrier
    
    lead1 = kwant.Builder(kwant.TranslationalSymmetry((-1,)))
    
    lead1[lat(0)] = onsite_normal
    lead1[lat(1), lat(0)] = hop
    
    lead2 = kwant.Builder(kwant.TranslationalSymmetry((1,)))
    
    lead2[lat(0)] = onsite_sc
    lead2[lat(1), lat(0)] = hop
    
    sys.attach_lead(lead1)
    sys.attach_lead(lead2)
    
    sys = sys.finalized()
    
    phlead.make_ph_lead(sys, 0, np.kron(sigmay, sigmay))
    
    return sys

def get_tunnel_spectroscopy(mu_l, mu_sc, Ez, alpha, delta, Vbarrier):
    sys = make_system_spectroscopy()
    Es = np.linspace(-0.15, 0.15, 101) + 1e-5
    Gs = []

    for E in Es:
        smat = kwant.smatrix(sys, energy=E, args=(mu_l, mu_sc, Ez, alpha, delta, Vbarrier))
        Gs.append(phlead.conductance(smat, 0))

    return Es,Gs


def plot_spectroscopy(Vbarrier):
    fig, ax = plt.subplots()

    # Trivial
    Es, Gs = get_tunnel_spectroscopy(0.5, 0.0, 0., 0.15, 0.1, Vbarrier)
    ax.plot(Es, Gs, 'k-',label = 'trivial')
    
    # Non Trivial
    Es, Gs = get_tunnel_spectroscopy(0.5, 0.0, 0.25, 0.15, 0.1, Vbarrier)
    ax.plot(Es, Gs, 'r-',label = 'non trivial')
    
    ax.set_xticks([ -0.1, 0.0, 0.1])
    ax.set_title('Tunnel spectroscopy, $V_{Barrier} = %1.2f $' %Vbarrier)
    ax.set_xlabel('Bias Voltage')
    ax.set_xlim([-0.15, 0.15])
    ax.set_ylabel('$G/G_0$')
    ax.set_ylim([-0.0,2.1])
    ax.set_yticks([0.0, 0.5, 1.0, 1.5, 2.0])
    
    ax.legend(fontsize=7)
       
    return fig

# nanowire ring
def nanowire_ring(Flux, N=100, mu=0.4, t=1.0, alpha=0.1, Delta=0.1, B=1.0):
    
    lat = kwant.lattice.chain()
    sys = kwant.Builder()
    
    sigma_0=np.array([[1, 0],[0,1]])
    sigma_x=np.array([[0,1],[1,0]])
    sigma_y=np.array([[0, -1j],[1j, 0]])
    sigma_z=np.array([[1,0],[0,-1]])
    

    onsite = ((mu -2*t)*np.kron(sigma_z, sigma_0)
              + B*np.kron(sigma_0,sigma_z)
              + Delta*np.kron(sigma_x,sigma_0))
    
    hopping = -(t*np.kron(sigma_z, sigma_0)+1j*alpha*np.kron(sigma_z,sigma_x))
        
    for n in xrange(N):
        sys[lat(n)] = onsite 
        
    for n in xrange(N-1):
        sys[lat(n), lat(n + 1)] = hopping
    temp = np.kron(np.array([[np.exp(1j*Flux/2), 0], [0, np.exp(-1j*Flux/2)]]), sigma_0)
    
    sys[lat(N-1), lat(0)] = 0.7*temp.dot(hopping)
    
    ham = sys.finalized().hamiltonian_submatrix()
    return ham


def get_nanowire_spectrum(trivial=False):
    if trivial:
        B=0.2
    else:
        B=1.0
        
    spectrum = []
    Fluxes = np.linspace(0, 2.0, 201)
    for Flux in Fluxes:
        H = nanowire_ring(2*np.pi*Flux, B=B)
        energies = np.linalg.eigvalsh(H)
        spectrum.append(energies)
    spectrum = np.array(spectrum)
    
    
    interesting = spectrum[:, np.shape(spectrum)[1]/2-1:np.shape(spectrum)[1]/2+1]
    first  = []
    second = []

    if trivial:
        first = interesting[:,0]
        second = interesting[:,1]
    else:
        for flux, E in zip(Fluxes,interesting):
            if flux < 0.5 or flux > 1.5:
                first.append(E[0])
                second.append(E[1])
            else:
                first.append(E[1])
                second.append(E[0])
            
    return Fluxes, spectrum, first, second


def plot_spectrum_nanowire(Fluxes, spectrum, first, second, ylim=[-0.2, 0.2]):
    fig, ax = plt.subplots()
    
    ax.plot(Fluxes, spectrum, '-', color='0.5', linewidth = 0.6)
    ax.plot(Fluxes, first, 'k-')
    ax.plot(Fluxes, second, 'r-')
    
    ax.axhline(0., ls=':', linewidth=1.2, c='k')
    ax.set_ylabel('$E$')
    ax.set_xlabel(r'$\Phi/\Phi_0$')
    ax.set_xticks([0, 1, 2])
    ax.set_ylim(ylim)
    ax.set_title('Spectrum')
    
def plot_gse_sc_nanowire(Fluxes, spectrum):
    first_lower_excited_index = np.shape(spectrum)[1]/2 - 1
    
    energy_gs = []
    for (spec, e) in zip(spectrum,first):
        energy_gs.append(np.sum(spec[:first_lower_excited_index])+e)

    energy_gs = energy_gs - max(energy_gs)
    current = np.diff(energy_gs)*len(energy_gs)


    fig = plt.figure(figsize=(9, 3.5))
    ax1 = fig.add_subplot(1, 2, 1)
    ax1.plot(Fluxes, energy_gs)
    ax1.set_xlabel('$\Phi/\Phi_0$')
    ax1.set_xticks([0, 1, 2])
    ax1.set_yticks([-0.06, -0.03, 0.0])
    ax1.set_ylim(-0.07,0.01)
    ax1.set_ylabel('$E_{gs}$')
    ax1.set_title('Ground state energy of the ring')

    ax2 = fig.add_subplot(1, 2, 2)
    ax2.plot((Fluxes[1:]+Fluxes[:len(Fluxes)-1])/2, current)
    ax2.set_xlabel('$\Phi/\Phi_0$')
    ax2.set_xticks([0, 1, 2])
    ax2.set_yticks([-0.2, 0.0, 0.2])
    ax2.set_ylabel('$I_{sc}$')
    ax2.set_title('Induced supercurrent')
    #plt.subplots_adjust(left = 0.15)
    fig.tight_layout()
    plt.show()  
```

# Majorana resonance in tunneling spectroscopy

Carlo Beenakker just introduced to you two important signatures of Majoranas, the $4\pi$-periodic Josephson effect and the resonant conductance peak, as two sides of the same coin. So let's start by studying one of the sides, the resonant peak.

To understand this signature we first have to understand how charge is transferred from a metallic lead to a superconductor. In general this transfer takes place via a mechanism known as Andreev reflection. Before we venture into the discussion of the conductance signatures of a Majorana zero mode, it will be useful to learn what Andreev reflection is.

## Andreev reflection

Let's consider the following, very simple circuit with two electrodes:

![](figures/ns_interface.png)

One electrode is a normal metal, the other a superconductor, and they are kept at a voltage difference $V$. At the interface between the normal metal and the superconductor (in short, NS interface) there is a barrier. Let's imagine that the potential difference is very small, in particular that $eV \ll \Delta$, the energy gap in the superconductor ($e$ is the charge of the electron). This small voltage difference pushes electrons from the normal metal towards the superconductor. The electrons have a kinetic energy of about $E_F$, the Fermi level in the metal. What happens when they arrive at the interface with superconductor? The superconductor has no electronic states available up to an energy $\Delta$ above the Fermi level, and the voltage is not enough to provide for this energy difference. How can a current develop?

To understand this, let's look more closely at an electron arriving at the interface with the superconductor. There are two possible processes that can take place, *normal reflection* and *Andreev reflection*. In normal reflection, the electron is reflected back by the potential barrier at the interface, even before attempting entering the superconductor:

![](figures/normal_reflection.png)

With normal reflection, there is no net charge transfer from the left electrode to the right electrode. Hence this process does not contribute any net current. In fact, the superconductor does not even participate directly in normal reflection, which would take place also if the right electrode was also in the normal metal state.

Andreev reflection, instead, is unique to the NS interface. In Andreev reflection, an electron is converted into a hole by the superconductor, and a Cooper pair is created in the superconductor.

![](figures/andreev_reflection.png)

You can see that a net charge of $2e$ is transferred from the left to the right electrode, hence Andreev reflection is the process responsible for the electrical current. (In principle, the simple transmission of an incident electron into the superconductor also contributes to the current. However, as we already mentioned, for us the voltage is so small that the transmission process can be ignored.)

You can think of Andreev reflection also as a *transmission* problem. Because of the presence of the superconductor, both electron and holes particupate in the transfer of charge in the normal metal lead. Conceptually, you can imagine to separate the left lead into two leads, one only carrying electrons and one only carrying holes. These two leads are connected by the superconductors, which converts incoming electrons in the first lead into outgoing holes in the second lead, and viceversa:

![](figures/andreev_as_transmission.png)

With this picture, you can understand that Andreev reflection is very similar to the problem of transmission through a double barrier.

## Andreev reflection off a Majorana zero mode

Now that we have understood a conventional NS interface, let's see what happens if our superconductor is topological:

![](figures/ns_majorana_interface.png)

You can imagine that the superconducting electrode is now a nanowire in the topological phase, like the one you have just studied. Because the superconductor is topological, there is a Majorana mode at the NS interface, whose wave function will “leak” a bit into the normal metal, through the barrier. Of course, there will be also a second Majorana mode, but we can imagine that it is located far enough from the NS interface, so that it does not have a role in the transport. Does the Majorana zero mode at the interface change the Andreev reflection properties?

Yes, and in a rather drastical way in fact. Going back to the picture of Andreev reflection as a transmission process through a double barrier, the crucial difference is that the Majorana mode now appears as a bound state between the two barriers:

![](figures/resonant_transmission_through_majorana.png)

In the double barrier problem in quantum mechanics, you can have **resonant** transmission in the presence of a bound state. This means that the probability to pass through the barriers can become one if the energy of the incident electron matches the energy of the bound state. In our case, the energy of the incident electron is $V$, and the energy of the bound state, the Majorana mode, is zero. So the presence of the Majorana mode leads to a **resonant peak** at $V=0$ for the tunneling of electrons into the topological nanowire.

This is perhaps the most direct way to measure the presence of a Majorana zero mode bound at the NS interface.


## Reflection matrix of a normal metal-superconductor interface

But let's go a bit beyond drawing cartoons. Quantum-mechanically, we can describe the situation in terms of a scattering problem. An incoming wave function $\Psi_\textrm{in}$ propagates in the left electrode, until it is reflected back at the interface with the superconductor, turning into an outgoing wave function $\Psi_\textrm{out}$. Because of the presence of the superconductor, both the incoming and outgoing wave have an electron-like and a hole-like component, exactly like the eigenstates of a Bogoliubov-de Gennes Hamiltonian: $\Psi_\textrm{in} = (\psi^e_\textrm{in}, \psi^h_\textrm{in})^T$, and similarly for $\Psi_\textrm{out}$.

Because there are no propagating waves in the right electrodes, this reflection process which relates $\Psi_\textrm{out}$ and $\Psi_\textrm{in}$ is unitary. So we have the following relation between incoming and ingoing wave:

$$\Psi_\textrm{out} = r \,\Psi_\textrm{in}\,$$

$$r=\left(\begin{array}{cc}r_{ee}&r_{eh}\\r_{he}&r_{hh}\end{array}\right).$$

The unitary matrix $r$ is known as the **reflection matrix**. Its complex elements are the amplitudes of normal and Andreev reflection of an incoming electron -- $r_{ee}$ and $r_{eh}$ -- and normal and Andreev reflection of an incoming hole -- $r_{hh}$ and $r_{he}$. The reflection matrix depends on the energy of the incoming wave (which is the same as that of the outgoing wave, since energy is conserved). In all that follows, we are always interested in the limit $V\to 0$, so we will only be concerned about the reflection matrix at the Fermi level. As usual, we measure all energies with respect to $E_F$, so we are concerned with the reflection matrix at zero energy.

Particle-hole symmetry of the excitations at zero-energy implies that we can replace $e\leftrightarrow h$ so that $r_{hh}^*=r_{ee}$ and $r_{eh}=r_{he}^*$. (**TO DO: PERHAPS EXPAND THIS PASSAGE BETTER? SYMMETRIES OF SCATTERING MATRICES ALWAYS CONFUSING IMHO **). In other words, we have the following symmetry of the reflection matrix:

$$\tau_x r^*\tau_x=r.$$

Once we know the reflection matrix of the interface, we can compute the conductance $G(V)$, which relates the current $I$ which develops as a response to a small voltage $V$, $G(V) = dI/dV$. The conductance is given by the following formula:

$$G(V)=2G_0|r_{eh}|^2.$$

We will not derive this equation, since it can be understood intuitively. The condutance is proportionally to the probability $|r_{eh}|^2$ of Andreev reflection, since we know that this is the only process that transfers electric charge from the left to the right electrode. The factor of $2$ reflects the fact that each Andreev reflection transfers a charge of a Cooper pair, $2e$. Finally, $G_0=e^2/h$ is the **conductance quantum**, the fundamental proportionality constant which relates currents to voltages.

To calculate the reflection probability $|r_{eh}|^2$, we must consider the BdG Hamiltonian of the system in Nambu space, as before. In this representation, the electrons and holes parts of the metallic lead  are effectively two separate leads that are connected through Andreev reflection in the superconductor. Thus, in this Nambu representation, the Andreev reflection process maps to a conventional impurity scattering problem from an electron lead into a hole lead through a double barrier transmission problem. 
The barrier is determined by the weak tunneling element $t'$ connecting the lead to the bound state.

The conductance resonance associated with Majorana modes is not uniquely topological because tunneling into any low energy bound state could produce such a resonance. One feature you might note from the numerical result is that the conductance appears to be fixed to $G=2G_0$ at $V=0.0$ independent of the barrier height. This turns out to be a topological feature of Majorana modes.

To see this let us examine the reflection matrix $r_{eh}$ in a little more detail for the case of a spin-polarized lead where only $\sigma_z=+1$ electrons and holes. The reflection matrix $r$, which includes both normal and Andreev reflection of electrons and holes in the normal lead is written as a $2\times 2$ matrix $$r=\left(\begin{array}{cc}r_{ee}&r_{eh}\\r_{he}&r_{hh}\end{array}\right).$$ Particle-hole symmetry of the excitations at zero-energy implies that we can replace $e\leftrightarrow h$ so that $r_{hh}^*=r_{ee}$ and $r_{eh}=r_{he}^*$. This constraint is written as the particle-hole symmetry of reflection matrices i.e. $$\tau_x r^*\tau_x=r.$$ In addition, since all electrons are reflected back into the lead (since there is no transmission), the 
reflection matrix $r$ is also unitary. These symmetry constraints mathematically characterize the reflection matrix in this problem.

Turns out that reflection matrices $r$ with the symmetry constraints above are also topological in their own way. To see this consider the quantity $Q=Det[\tau_x r]$, which is real by particle-hole symmetry: 
$$Q=Det[\tau_x r]=Det[r^*\tau_x]=Det[\tau_x r]^*=Q^*.$$ Since $r$ is unitary $|Q|=|Det[\tau_x r]|=1$, which means that $Q$ can only take discrete values $$Q=\pm 1.$$ Thus $Q$ cannot change as long as $r$ is unitary, which is true as long as the gap in the superconductor doesn't change. This is an example of another kind of topological invariant of a scattering topological invariant.

For $2\times 2$ matrices $r$, $Q=-Det[r]=|r_{ee}|^2-|r_{eh}|^2=\pm 1$. Additionally, unitarity requires $|r_{ee}|^2+|r_{eh}|^2=1$ so that either $|r_{ee}|=1$ (perfect reflection) or $|r_{eh}|=1$ perfect Andreev reflection. The situation cannot change without a phase transition. Thus the quantized conductance of the Majorana mode is topologically robust in this case and survives past the tunneling limit.

# Josephson effect in a topological nanowire Josephson junction is anomalously 4$\pi$ periodic

* As we saw in 1.2 changing the sign of the hopping across a bond of a topological superconducting wire results in a fermion parity change between the initial and final Hamiltonians. The sign change in the hopping might also be obtained by introducing a phase in the nanowire model as $$t_{1,N}\rightarrow t_{1,N}e^{i \theta/2},$$where $\theta$ is a phase twist that may be introduced by a vector potential. For a system in a ring, the vector potential is set by the magnetic flux $\Phi$ through the ring according to $$\theta=4\pi \Phi/\Phi_0,$$ where $\Phi_0=hc/2e$ is the superconducting flux quantum. 


* The fermion parity change with each flux increase $\Phi\rightarrow (\Phi+\Phi_0)$ leads to a level crossing and a $2\Phi_0$ periodicity in the energy levels with fixed fermion parity. 

Figure/Code: Plot spectrum of nanowire as a function of $\Phi$. 

* The level crossing is a local process and persists in an open wire geometry as well. In such a geometry, the hopping phase shift in $t_{1,N}$ ,may be removed from the Hamiltonian by a "gauge" transformation $c_L^\dagger \rightarrow c_L^\dagger e^{-i\theta/2}$. The same gauge transformation transforms the pairing term in the  Hamiltonian as $$\Delta_L\rightarrow \Delta_L e^{i\theta},$$ 
which is equivalent to a change in the superconducting phase by $\theta$. Such a system where the superconducting phase changes is referred to as a Josephson junction.

* While the BdG spectrum of a Josephson junction $H(\theta)$ is $2\pi$-periodic in $\theta$, as noted for the periodic boundary condition, in the topological case, the occupation is $4\pi$ periodic in $\theta$. Thus, the supercurrent $$I(\theta)=\partial_{\theta}E_{tot}(\theta),$$ where $E_{tot}(\theta)$ 
is the total energy of the junction. 

Figure/code: plot total energy and current as a function of flux.


Such a $4\pi$ Josephson effect, which may be measured experimentally is very unusual since without the fermion-parity induced occupation change, the Josephson current must be $2\pi$ periodic. In fact, there is a theorem that the Josephson current is $2\pi$ periodic in the grand canonical ensemble (i.e. the particle number is allowed to vary)

```python
spectrum = []
Fluxes = np.linspace(0, 2.0, 201)
for Flux in Fluxes:
    H = nanowire_ring(100, 0.1, 1.0, 0.1, 0.1, 0.2, 2*np.pi*Flux)
    energies = np.linalg.eigvalsh(H)
    spectrum.append(energies)


plot_spectrum(Fluxes, spectrum, [-0.11, 0.11])
```
