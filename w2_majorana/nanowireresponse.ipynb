```python
import phlead
sigma0 = np.array([[1., 0.], [0., 1.]])
sigmax = np.array([[0., 1.], [1., 0.]])
sigmay = np.array([[0., -1j], [1j, 0.]])
sigmaz = np.array([[1., 0.], [0., -1.]])



### Tunel spectroscopy
def make_system_spectroscopy(t=1):
    def onsite_sc(site, mu_l, mu_sc, Ez, alpha, delta, Vbarrier):
        return (2*t - mu_sc) * np.kron(sigma0, sigmaz) + Ez * np.kron(sigmax, sigma0) + delta * np.kron(sigma0, sigmax)

    def onsite_normal(site, mu_l, mu_sc, Ez, alpha, delta, Vbarrier):
        return (2*t - mu_l) * np.kron(sigma0, sigmaz) + Ez * np.kron(sigmax, sigma0)
    
    def onsite_barrier(site, mu_l, mu_sc, Ez, alpha, delta, Vbarrier):
        return (2*t - mu_l + Vbarrier) * np.kron(sigma0, sigmaz) + Ez * np.kron(sigmax, sigma0) 
    
    def hop(site1, site2, mu_l, mu_sc, Ez, alpha, delta, Vbarrier):
        return -t * np.kron(sigma0, sigmaz) + 1j * alpha * np.kron(sigmay, sigmaz)
    
    lat = kwant.lattice.chain()
    sys = kwant.Builder()
    
    # The system just consists of the tunnel barrier
    sys[lat(0)] = onsite_barrier
    
    lead1 = kwant.Builder(kwant.TranslationalSymmetry((-1,)))
    
    lead1[lat(0)] = onsite_normal
    lead1[lat(1), lat(0)] = hop
    
    lead2 = kwant.Builder(kwant.TranslationalSymmetry((1,)))
    
    lead2[lat(0)] = onsite_sc
    lead2[lat(1), lat(0)] = hop
    
    sys.attach_lead(lead1)
    sys.attach_lead(lead2)
    
    sys = sys.finalized()
    
    phlead.make_ph_lead(sys, 0, np.kron(sigmay, sigmay))
    
    return sys

def get_tunnel_spectroscopy(mu_l, mu_sc, Ez, alpha, delta, Vbarrier):
    sys = make_system_spectroscopy()
    Es = np.linspace(-0.15, 0.15, 101) + 1e-5
    Gs = []

    for E in Es:
        smat = kwant.smatrix(sys, energy=E, args=(mu_l, mu_sc, Ez, alpha, delta, Vbarrier))
        Gs.append(phlead.conductance(smat, 0))

    return Es,Gs


def plot_spectroscopy(Vbarrier):
    fig, ax = plt.subplots()

    # Trivial
    Es, Gs = get_tunnel_spectroscopy(0.5, 0.0, 0., 0.15, 0.1, Vbarrier)
    ax.plot(Es, Gs, 'k-',label = 'trivial')
    
    # Non Trivial
    Es, Gs = get_tunnel_spectroscopy(0.5, 0.0, 0.25, 0.15, 0.1, Vbarrier)
    ax.plot(Es, Gs, 'r-',label = 'non trivial')
    
    ax.set_xticks([ -0.1, 0.0, 0.1])
    ax.set_title('Tunnel spectroscopy, $V_{Barrier} = %1.2f $' %Vbarrier)
    ax.set_xlabel('Bias Voltage')
    ax.set_xlim([-0.15, 0.15])
    ax.set_ylabel('$G/G_0$')
    ax.set_ylim([-0.0,2.1])
    ax.set_yticks([0.0, 0.5, 1.0, 1.5, 2.0])
    
    ax.legend(fontsize=7)
       
    return fig

# nanowire ring
def nanowire_ring(Flux, N=100, mu=0.4, t=1.0, alpha=0.1, Delta=0.1, B=1.0):
    
    lat = kwant.lattice.chain()
    sys = kwant.Builder()
    
    sigma_0=np.array([[1, 0],[0,1]])
    sigma_x=np.array([[0,1],[1,0]])
    sigma_y=np.array([[0, -1j],[1j, 0]])
    sigma_z=np.array([[1,0],[0,-1]])
    

    onsite = ((mu -2*t)*np.kron(sigma_z, sigma_0)
              + B*np.kron(sigma_0,sigma_z)
              + Delta*np.kron(sigma_x,sigma_0))
    
    hopping = -(t*np.kron(sigma_z, sigma_0)+1j*alpha*np.kron(sigma_z,sigma_x))
        
    for n in xrange(N):
        sys[lat(n)] = onsite 
        
    for n in xrange(N-1):
        sys[lat(n), lat(n + 1)] = hopping
    temp = np.kron(np.array([[np.exp(1j*Flux/2), 0], [0, np.exp(-1j*Flux/2)]]), sigma_0)
    
    sys[lat(N-1), lat(0)] = 0.7*temp.dot(hopping)
    
    ham = sys.finalized().hamiltonian_submatrix()
    return ham


def get_nanowire_spectrum(trivial=False):
    if trivial:
        B=0.2
    else:
        B=1.0
        
    spectrum = []
    Fluxes = np.linspace(0, 2.0, 201)
    for Flux in Fluxes:
        H = nanowire_ring(2*np.pi*Flux, B=B)
        energies = np.linalg.eigvalsh(H)
        spectrum.append(energies)
    spectrum = np.array(spectrum)
    
    
    interesting = spectrum[:, np.shape(spectrum)[1]/2-1:np.shape(spectrum)[1]/2+1]
    first  = []
    second = []

    if trivial:
        first = interesting[:,0]
        second = interesting[:,1]
    else:
        for flux, E in zip(Fluxes,interesting):
            if flux < 0.5 or flux > 1.5:
                first.append(E[0])
                second.append(E[1])
            else:
                first.append(E[1])
                second.append(E[0])
            
    return Fluxes, spectrum, first, second


def plot_spectrum_nanowire(Fluxes, spectrum, first, second, ylim=[-0.2, 0.2]):
    fig, ax = plt.subplots()
    
    ax.plot(Fluxes, spectrum, '-', color='0.5', linewidth = 0.6)
    ax.plot(Fluxes, first, 'k-')
    ax.plot(Fluxes, second, 'r-')
    
    ax.axhline(0., ls=':', linewidth=1.2, c='k')
    ax.set_ylabel('$E$')
    ax.set_xlabel(r'$\Phi/\Phi_0$')
    ax.set_xticks([0, 1, 2])
    ax.set_ylim(ylim)
    ax.set_title('Spectrum')
    
def plot_gse_sc_nanowire(Fluxes, spectrum):
    first_lower_excited_index = np.shape(spectrum)[1]/2 - 1
    
    energy_gs = []
    for (spec, e) in zip(spectrum,first):
        energy_gs.append(np.sum(spec[:first_lower_excited_index])+e)

    energy_gs = energy_gs - max(energy_gs)
    current = np.diff(energy_gs)*len(energy_gs)


    fig = plt.figure(figsize=(9, 3.5))
    ax1 = fig.add_subplot(1, 2, 1)
    ax1.plot(Fluxes, energy_gs)
    ax1.set_xlabel('$\Phi/\Phi_0$')
    ax1.set_xticks([0, 1, 2])
    ax1.set_yticks([-0.06, -0.03, 0.0])
    ax1.set_ylim(-0.07,0.01)
    ax1.set_ylabel('$E_{gs}$')
    ax1.set_title('Ground state energy')

    ax2 = fig.add_subplot(1, 2, 2)
    ax2.plot((Fluxes[1:]+Fluxes[:len(Fluxes)-1])/2, current)
    ax2.set_xlabel('$\Phi/\Phi_0$')
    ax2.set_xticks([0, 1, 2])
    ax2.set_yticks([-0.2, 0.0, 0.2])
    ax2.set_ylabel('$I_{sc}$')
    ax2.set_title('Current')
    #plt.subplots_adjust(left = 0.15)
    fig.tight_layout()
    plt.show()  
```

# Majorana resonance in tunneling spectroscopy

Carlo Beenakker just introduced to you two important signatures of Majoranas, the $4\pi$-periodic Josephson effect and the resonant conductance peak, as two sides of the same coin. So let's start by studying one of the sides, the resonant peak.

To understand this signature we first have to understand how charge is transferred from a metallic lead to a superconductor. In general this transfer takes place via a mechanism known as Andreev reflection. Before we venture into the discussion of the conductance signatures of a Majorana zero mode, it will be useful to learn what Andreev reflection is.

## Andreev reflection

Let's consider the following, very simple circuit with two electrodes:

![](figures/ns_interface.png)

One electrode is a normal metal, the other a superconductor, and they are kept at a voltage difference $V$. At the interface between the normal metal and the superconductor (in short, NS interface) there is a barrier. We are particularly interested in the case when the voltage difference is very small, namely $eV \ll \Delta$, the energy gap in the superconductor ($e$ is the charge of the electron). This small voltage difference pushes electrons from the normal metal towards the superconductor. The electrons have a kinetic energy of about $E_F$, the Fermi level in the metal. What happens when they arrive at the interface with superconductor? The superconductor has no electronic states available up to an energy $\Delta$ around the Fermi level, and the voltage is not enough to provide for this energy difference. How can a current develop?

To understand this, let's look more closely at an electron arriving at the interface with the superconductor. There are two possible processes that can take place, *normal reflection* and *Andreev reflection*. In normal reflection, the electron is simply reflected at the interface with the superconductor:

![](figures/normal_reflection.png)

With normal reflection, there is no net charge transfer from the left electrode to the right electrode. Hence this process does not contribute any net current. In fact, the superconductor does not even participate directly in normal reflection, which would take place also if the right electrode was also in the normal metal state.

Andreev reflection, instead, is unique to the NS interface. In Andreev reflection, an electron is converted into a hole by the superconductor, and a Cooper pair is created in the superconductor.

![](figures/andreev_reflection.png)

You can see that a net charge of $2e$ is transferred from the left to the right electrode, hence Andreev reflection is the process responsible for the electrical current. (In principle, the simple transmission of an incident electron into the superconductor also contributes to the current. However, as we already mentioned, for us the voltage is so small that the transmission process can be ignored.)

You can think of Andreev reflection also as a *transmission* problem. Because of the presence of the superconductor, both electron and holes particupate in the transfer of charge in the normal metal lead. Conceptually, you can imagine to separate the left lead into two leads, one only carrying electrons and one only carrying holes. These two leads are connected by the superconductor, which converts incoming electrons in the first lead into outgoing holes in the second lead, and viceversa:

![](figures/andreev_as_transmission.png)

With this picture, you can understand that Andreev reflection is very similar to the problem of transmission through a double barrier.

Let's call $r_{eh}$ the amplitude for Andreev reflection. Its absolute value squared, $\left|r_{eh}\right|^2$, is the probability that an incoming electron from the normal metal is Andreev reflected as a hole. Once we know $r_{eh}$, we can compute the conductance $G(V)$, which relates the current $I$ which develops as a response to a small voltage $V$, $G(V) = dI/dV$. The conductance is given by the following formula:

$$G(V)=2G_0|r_{eh}|^2.$$

We will not derive this equation, since it can be understood intuitively. The condutance is proportional to the probability $|r_{eh}|^2$ of Andreev reflection, since we know that this is the only process that transfers electric charge from the left to the right electrode. The factor of $2$ reflects the fact that each Andreev reflection transfers a charge of a Cooper pair, $2e$. Finally, $G_0=e^2/h$ is the **conductance quantum**, the fundamental proportionality constant which relates currents to voltages.

## Andreev reflection off a Majorana zero mode

Now that we have understood a conventional NS interface, let's see what happens if our superconductor is topological:

![](figures/ns_majorana_interface.png)

You can imagine that the superconducting electrode is now a nanowire in the topological phase, like the one you have just studied. Because the superconductor is topological, there is a Majorana mode at the NS interface, whose wave function will “leak” a bit into the normal metal, through the barrier. Of course, there will be also a second Majorana mode, but we can imagine that it is located far enough from the NS interface, so that it does not have a role in the transport. Does the Majorana zero mode at the interface change the Andreev reflection properties?

Yes, and in a rather drastical way in fact. Going back to the picture of Andreev reflection as a transmission process through a double barrier, the crucial difference is that the Majorana mode now appears as a bound state between the two barriers:

![](figures/resonant_transmission_through_majorana.png)

In the double barrier problem in quantum mechanics, you can have **resonant** transmission in the presence of a bound state. This means that the probability $\left|r_{eh}\right|^2$ to pass through the barriers is dramatically enhanced if the energy of the incident electron matches the energy of the bound state. In our case, the energy of the incident electron is $V$, and the energy of the bound state, the Majorana mode, is zero. So the presence of the Majorana mode leads to a **resonant peak** in the conductance of the NS interface at $V=0$.

This is perhaps the most direct way to measure the presence of a Majorana zero mode at the end of a topological superconductor.
However, the conductance resonance associated with Majorana modes is not uniquely topological, because tunneling into any low energy bound state could produce such a resonance. Is there anything in particular which distinguishes the Majorana resonance from any other resonance? To answer this question, we must go a bit beyond drawing cartoons.

By looking at the resonant peak in the plot above, you will certainly notice a very robust and persistent characteristic:
its height is quantized to the value of $2G_0$, independently of the strength of the voltage barrier. From the formula above, this means that if a Majorana is present we have $\left|r_{eh}\right|^2=1$, that is we have **perfect Andreev reflection**. In the following, we will show how this quantized value of the conductance follows from topological considerations.

## Reflection matrix of a normal metal-superconductor interface

Quantum-mechanically, we can describe transport through the NS interface as a scattering problem. An incoming wave function $\Psi_\textrm{in}$ propagates in the left electrode, until it is reflected back at the interface with the superconductor, turning into an outgoing wave function $\Psi_\textrm{out}$. Because of the presence of the superconductor, both the incoming and outgoing wave have an electron-like and a hole-like component, exactly like the eigenstates of a Bogoliubov-de Gennes Hamiltonian: $\Psi_\textrm{in} = (\psi^e_\textrm{in}, \psi^h_\textrm{in})^T$, and similarly for $\Psi_\textrm{out}$.

So we have the following relation between incoming and ingoing wave:

$$\Psi_\textrm{out} = r(V) \,\Psi_\textrm{in}\,$$

$$r(V)=\left(\begin{array}{cc}r_{ee}&r_{eh}\\r_{he}&r_{hh}\end{array}\right).$$

The matrix $r$ is known as the **reflection matrix**. Its complex elements are the amplitudes of normal and Andreev reflection of an incoming electron - $r_{ee}$ and $r_{eh}$ - and normal and Andreev reflection of an incoming hole - $r_{hh}$ and $r_{he}$. These amplitudes all depend on $V$, but we did not write that explicitly.

Because for $eV\ll \Delta$ there are no propagating waves in the superconductor, the reflection process which relates $\Psi_\textrm{out}$ and $\Psi_\textrm{in}$ is unitary, $r^\dagger r=1$. This implies that

$$\left|r_{ee}\right|^2+\left|r_{eh}\right|^2 = \left|r_{he}\right|^2+\left|r_{hh}\right|^2 = 1\,.$$

This is the mathematical way of saying that an electron (or hole) arriving at the interface has no alternatives other than being normal-reflected or Andreev-reflected.

Can we add any other constraint to $r$, that might be insightful to distinguish any characteristic of the Majorana mode? Just like we did last week, let's try to study $r$ using symmetry and topology. Our circuit involves a superconductor, so we must have particle-hole symmetry in the problem. In fact, in order to derive $r$ explicitly we could start directly from the Bogoliubov-de Gennes Hamiltonian of the NS system, and solve it for an energy $V$. This is a lot of work, which we won't do. It is more interesting to understand what are the consequences of particle-hole symmetry directly on $r$.

First, particle-hole symmetry exchanges electrons and hole components of the wave function, so it involves a Pauli matrix $\tau_x$ acting on $\Psi_\textrm{in}$ or $\Psi_\textrm{out}$. Second, it is an anti-unitary symmetry, so it involves complex conjugation. Third, it changes the sign of the energy so it sends $V$ into $-V$. Hence we arrive at the following symmetry on the reflection matrix:

$$\tau_x r^*(-V) \tau_x = r(V)\,.$$

Together with unitarity, particle-hole symmetry imposes that the conductance is symmetric around zero voltage, $G(V)=G(-V)$. At $V=0$, the case which most interests us because there is where we expect a Majorana resonance, we have:

$$ \tau_x r^*_0 \tau_x = r_0 \,.$$

where we defined $r_0\equiv r(V=0)$. So much for the impact of symmetry on $r$. What about topology?

## Topological invariant of the reflection matrix

The Majorana zero mode is the consequence of a topological phase in the topological superconductor, and their presence is dictated by the bulk-boundary correspondence. Can we find any consequence on this fact on $r_0$? It turns out that reflection matrices $r$ with particle-hole symmetry are also topological in their own way. Their topological invariant is 

$$Q = \det\,r_0\,.$$

Again, we will not *derive* this equation from first principles, but rather convince you that the determinant on the right side has all the characteristics that a topological invariant should have.

First of all, the determinant of a unitary matrix such as $r_0$ is always a complex number with unit norm, so $\left|\det\,r_0\,\right|=1$. Second, because of particle-hole symmetry, the determinant is real: $\det r_0 = \det\, (\tau_x r^*_0\,\tau_x) = \det\,r_0^*\,=(\det\,r_0)^*$. Hence, $\det\,r_0\,= \pm 1$. This is quite promising! Two possible discrete values, just like the Pfaffian invariant of the Kitaev chain.

Because it is just dictated by unitarity and particle-hole symmetry, the determinant of $r$ cannot change from $+1$ to $-1$ under a change of the properties of the NS interface. For instance, you can very the height of the potential barrier at the interface, but this cannot affect the determinant of $r$.

The only way to make the determinant change sign is to close the bulk gap in the superconducting electrode. If the gap goes to zero, then it is not true that an incoming electron coming from the normal metal can only be normal-reflected or Andreev-reflected. It can also just enter the superconducting electrode as an electron. Hence the reflection matrix does not contain anymore all the possible processes taking place at the interface, and it won't be unitary anymore. This allows the determinant to change sign. We conclude that $Q=\det\,r$ is a good topological invariant.

Explicitly, we have that

$$Q=|r_{ee}|^2-|r_{eh}|^2\equiv\pm 1\,.$$

We already saw that unitarity requires that $|r_{ee}|^2+|r_{eh}|^2=1$. There are only two possibilities for both conditions to be true: either $|r_{ee}|=1$ (**perfect  normal reflection**) or $|r_{eh}|=1$ (**perfect Andreev reflection**). The situation cannot change without a phase transition. Thus the quantized conductance of the Majorana mode is topologically robust in this case and survives past the tunneling limit.

# $4\pi$-periodic Josephson effect

* As we saw in 1.2 changing the sign of the hopping across a bond of a topological superconducting wire results in a fermion parity change between the initial and final Hamiltonians. The sign change in the hopping might also be obtained by introducing a phase in the nanowire model as $$t_{1,N}\rightarrow t_{1,N}e^{i \theta/2},$$where $\theta$ is a phase twist that may be introduced by a vector potential. For a system in a ring, the vector potential is set by the magnetic flux $\Phi$ through the ring according to $$\theta=4\pi \Phi/\Phi_0,$$ where $\Phi_0=hc/2e$ is the superconducting flux quantum. 


* The fermion parity change with each flux increase $\Phi\rightarrow (\Phi+\Phi_0)$ leads to a level crossing and a $2\Phi_0$ periodicity in the energy levels with fixed fermion parity. 

Figure/Code: Plot spectrum of nanowire as a function of $\Phi$. 

* The level crossing is a local process and persists in an open wire geometry as well. In such a geometry, the hopping phase shift in $t_{1,N}$ ,may be removed from the Hamiltonian by a "gauge" transformation $c_L^\dagger \rightarrow c_L^\dagger e^{-i\theta/2}$. The same gauge transformation transforms the pairing term in the  Hamiltonian as $$\Delta_L\rightarrow \Delta_L e^{i\theta},$$ 
which is equivalent to a change in the superconducting phase by $\theta$. Such a system where the superconducting phase changes is referred to as a Josephson junction.

* While the BdG spectrum of a Josephson junction $H(\theta)$ is $2\pi$-periodic in $\theta$, as noted for the periodic boundary condition, in the topological case, the occupation is $4\pi$ periodic in $\theta$. Thus, the supercurrent $$I(\theta)=\partial_{\theta}E_{tot}(\theta),$$ where $E_{tot}(\theta)$ 
is the total energy of the junction. 

Figure/code: plot total energy and current as a function of flux.


Such a $4\pi$ Josephson effect, which may be measured experimentally is very unusual since without the fermion-parity induced occupation change, the Josephson current must be $2\pi$ periodic. In fact, there is a theorem that the Josephson current is $2\pi$ periodic in the grand canonical ensemble (i.e. the particle number is allowed to vary)
