# to calculate conductance from Andreev reflection into superconducting bound states

* Bound states in superconductors can be detected by tunneling electrons from a metallic lead. Metallic leads (particularly in Nambu space) can be thought of as supporting two types of carriers electrons $c^\dagger$ and holes $c$. These carriers are  Andreev reflection of an electron from a metallic lead into a hole in the lead results in current between the metallic lead and superconductor. The electrons and holes in the metallic lead may be thought of as two separate leads that are connected through Andreev reflection in the superconductor. The Andreev reflection process maps to a conventional impurity scattering problem from 
an electron lead into a hole lead through an impurity, with an additional particle-hole symmetry. Schematic figure: to illustrate this idea. On one hand, Andreev reflection  schematic at a normal/superconducting junction with a bound state in the superconductor on the other half - claim equivalence to a double barrier problem with an electron lead and a hole lead.

* In the electron-hole lead picture, Andreev reflection from an electron into a hole maps to transmission from the electron to the hole lead. Since each transmission event leads to a charge transfer of $2e$, the conductance is $$G(V)=\frac{2e^2}{h}|r_{eh}|^2,$$
where $r_{eh}$ is the element of the reflection matrix from the electron to the hole lead.


* Bound states that are weakly coupled to the lead result in resonant transmission peaks as with conventional transmission problems when the voltage $V$ matches the energy $E$ of the bound state.

* For topological nanowires, the zero-energy Majorana state leads to a resonance at $V=0$. The Majorana bound state is separated from the electron and hole leads by barriers potentials that are related by particle-hole symmetry. The equivalence of the electron and hole barriers the Majorana conductance problem maps to a symmetric double barrier problem, which demonstrates perfect transmission at zero energy with a width inversely related to the barrier height. The perfect transmission in the double barrier problem maps to perfect Andreev reflection which corresponds to $|r_{eh}|=1$ and a conductance $G(V=0)=2 e^2/h$.

# Conductance into Majorana  nanowire edge is quantized

* The complete process at zero-energy can be represented by a $2\times 2$ reflection matrix $r$ between the electron and hole leads.
The reflection matrix conserves probability so that $r$ is unitary. Particle-hole symmetry requires $\tau_x r^*\tau_x=r$ so that the determinant $$Q=Det[\tau_x r]=Det[r^*\tau_x]=Det[\tau_x r]^*=\pm 1,$$ since this means $Q$ is real and for unitary $r$ matrices $|Q|=|Det[r]|=1$. Clearly, similar to the Pfaffian topological invariant for the bulk $1D$ superconductor, $Q=\pm 1$ cannot 
change unless $S$ becomes non-unitary with $Det[S]=0$. The reflection matrix $r$ can only become non-unitary when the superconducting gap in the bulk closes. The topological invariant $Q=Det[\tau_x r(V=0)]$ is an example for a scattering topological invariant.
* For the simplest case of spin-polarized leads $r$ is a $2\times 2$ matrix so that particle-hole symmetry forces $$r=\left(\begin{array}{cc}a&b\\b^*&a^*\end{array}\right)$$ with $|a|^2-|b|^2=Q=\pm 1$. Additionally, unitarity requires $|a|^2+|b|^2=1$ so that either $|a|=1$ (perfect reflection) or $|b|=1$ perfect Andreev reflection. The situation cannot change without a phase transition. Thus the quantized conductance of the Majorana mode is topologically robust in this case and survives past the tunneling limit. Figure/code: Calculate  $V=0$ conductance and $Q$ and Pfaffian as a function of $B$ and see how they change together across the topological trnasition. 
* Particle-hole symmetry transforms the electron lead into the hole lead and vice-versa along with transforming $V\rightarrow -V$. Reciprocity of scattering i.e. $r_{eh}(V)=r_{he}^\dagger(-V)$ guarantees that the conductance is particle hole symmetric $G(V)=G(-V)$.

* For conventional pair of sub-gap particle-hole symmetric bound states at energy $\pm \epsilon$, the destructive interference between transmission from a spin-polarized lead through the positive and negative states cancels the conductance. Thus conductance vanishes strictly in the middle of the finite size splitting.


```python
from math import sqrt
```

```python
sigma0 = np.array([[1., 0.], [0., 1.]])
sigmax = np.array([[0., 1.], [1., 0.]])
sigmay = np.array([[0., -1j], [1j, 0.]])
sigmaz = np.array([[1., 0.], [0., -1.]])
```

```python
def make_system(t=1):
    def onsite_sc(site, mu_l, mu_sc, Ez, alpha, delta, Vbarrier):
        return (2*t - mu_sc) * np.kron(sigma0, sigmaz) + Ez * np.kron(sigmax, sigma0) + delta * np.kron(sigma0, sigmax)

    def onsite_normal(site, mu_l, mu_sc, Ez, alpha, delta, Vbarrier):
        return (2*t - mu_l) * np.kron(sigma0, sigmaz) + Ez * np.kron(sigmax, sigma0)
    
    def onsite_barrier(site, mu_l, mu_sc, Ez, alpha, delta, Vbarrier):
        return (2*t - mu_l + Vbarrier) * np.kron(sigma0, sigmaz) + Ez * np.kron(sigmax, sigma0) 
    
    def hop(site1, site2, mu_l, mu_sc, Ez, alpha, delta, Vbarrier):
        return -t * np.kron(sigma0, sigmaz) + 1j * alpha * np.kron(sigmay, sigmaz)
    
    lat = kwant.lattice.chain()
    sys = kwant.Builder()
    
    # The system just consists of the tunnel barrier
    sys[lat(0)] = onsite_barrier
    
    lead1 = kwant.Builder(kwant.TranslationalSymmetry((-1,)))
    
    lead1[lat(0)] = onsite_normal
    lead1[lat(1), lat(0)] = hop
    
    lead2 = kwant.Builder(kwant.TranslationalSymmetry((1,)))
    
    lead2[lat(0)] = onsite_sc
    lead2[lat(1), lat(0)] = hop
    
    sys.attach_lead(lead1)
    sys.attach_lead(lead2)
    
    sys = sys.finalized()
    
    phlead.make_ph_lead(sys, 0, np.kron(sigmay, sigmay))
    
    return sys
```

```python
def plot_G(mu_l, mu_sc, Ez, alpha, delta, Vbarrier):

    sys = make_system()

    Q = sqrt(mu_sc**2 + delta**2) - Ez

    Es = np.linspace(-0.2, 0.2, 101) + 1e-5
    Gs = []

    for E in Es:
        smat = kwant.smatrix(sys, energy=E, args=(mu_l, mu_sc, Ez, alpha, delta, Vbarrier))
        Gs.append(phlead.conductance(smat, 0))
    
    fig, ax = plt.subplots()
    ax.plot(Es, Gs)
    ax.set_xticks([-0.2, -0.1, 0.0, 0.1, 0.2])
    
    if Q > 0:
        ax.set_title('System is trivial, $B = %1.3f $' %Ez)
    else:
        ax.set_title('System is topological, $B = %1.3f $' %Ez)
    
    
    ax.set_xlabel('Bias Voltage')
    ax.set_ylabel('$G/G_0$')
    ax.set_ylim([-0.1,2.1])
    ax.set_yticks([0.0, 0.5, 1.0, 1.5, 2.0])
    return fig
```

```python
StaticInteract(lambda B, Vbarrier: plot_G(0.5, 0.1, B, 0.15, 0.05, Vbarrier),
         B=RangeWidget(0.0, 0.5+1e-7, 0.025, default=0.3, show_range=False), 
         Vbarrier=RangeWidget(1, 3, 1, default=2, show_range=True))
```

# Josephson effect in a topological nanowire Josephson junction is anomalously 4$\pi$ periodic

* As we saw in 1.2 changing the sign of the hopping across a bond of a topological superconducting wire results in a fermion parity change between the initial and final Hamiltonians. The sign change in the hopping might also be obtained by introducing a phase in the nanowire model as $$t_{1,N}\rightarrow t_{1,N}e^{i \theta/2},$$where $\theta$ is a phase twist that may be introduced by a vector potential. For a system in a ring, the vector potential is set by the magnetic flux $\Phi$ through the ring according to $$\theta=4\pi \Phi/\Phi_0,$$ where $\Phi_0=hc/2e$ is the superconducting flux quantum. 


* The fermion parity change with each flux increase $\Phi\rightarrow (\Phi+\Phi_0)$ leads to a level crossing and a $2\Phi_0$ periodicity in the energy levels with fixed fermion parity. Figure/Code: Plot spectrum of nanowire as a function of $\Phi$. 

* The level crossing is a local process and persists in an open wire geometry as well. In such a geometry, the hopping phase shift in $t_{1,N}$ ,may be removed from the Hamiltonian by a "gauge" transformation $c_L^\dagger \rightarrow c_L^\dagger e^{-i\theta/2}$. The same gauge transformation transforms the pairing term in the  Hamiltonian as $$\Delta_L\rightarrow \Delta_L e^{i\theta},$$ 
which is equivalent to a change in the superconducting phase by $\theta$. Such a system where the superconducting phase changes is referred to as a Josephson junction.

* While the BdG spectrum of a Josephson junction $H(\theta)$ is $2\pi$-periodic in $\theta$, as noted for the periodic boundary condition, in the topological case, the occupation is $4\pi$ periodic in $\theta$. Thus, the supercurrent $$I(\theta)=\partial_{\theta}E_{tot}(\theta),$$ where $E_{tot}(\theta)$ 
is the total energy of the junction. Figure/code: plot total energy and current as a function of flux.
Such a $4\pi$ Josephson effect, which may be measured experimentally is very unusual since without the fermion-parity induced occupation change, the Josephson current must be $2\pi$ periodic. In fact, there is a theorem that the Josephson current is $2\pi$ periodic in the grand canonical ensemble (i.e. the particle number is allowed to vary)
